```python
# Import dependencies

import numpy as np
import pandas as pd
import datetime as dt

# Statistical analysis
from scipy import stats
from numpy import mean

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
```

```python
# Using either SQLAlchemy or pandas, get the requested data per the README
# Create Database Connection

engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model

# reflect the tables

# Use the Inspector to explore the database and print the table names
# YOUR CODE HERE
inspector = inspect(engine)
inspector.get_table_names()
```

```python
# Use Inspector to print the column names and types
# YOUR CODE HERE
columns = inspector.get_columns('measurement')
for column in columns:
    print(column["name"], column["type"])
```

```python
# Use SQLAlchemy automap_base() to reflect your tables into classes and 
# save a reference to those classes called Station and Measurement.
# View all of the classes that automap found
# Reflect Database into ORM class

Base = automap_base()
Base.prepare(engine, reflect=True)
Measurement = Base.classes.measurement
```

```python
# Create our session (link) from Python to the DB
# Start a session to query the database

session = Session(engine)
```

```python
# Identify the average temperature in June at all stations across all available years in the dataset. 
# Do the same for December temperature.
date_str ="06"
june_temp_mean = session.query(func.avg(Measurement.tobs)).\
                    filter(func.strftime("%m", Measurement.date) == date_str).\
                    all()
                     
                    
june_temp_mean
```

```python
# Identify the average temperature in June at all stations across all available years in the dataset. 
# Do the same for December temperature.
date_str ="12"
dec_temp_mean = session.query(func.avg(Measurement.tobs)).\
                    filter(func.strftime("%m", Measurement.date) == date_str).\
                    all()
                     
                    
dec_temp_mean
```

```python
date_str ="06"
june_temp = session.query(Measurement.tobs).\
                    filter(func.strftime("%m", Measurement.date) == date_str).\
                    all()
                     
                    
june_temp
```

```python
date_str ="12"
dec_temp = session.query(Measurement.tobs).\
                    filter(func.strftime("%m", Measurement.date) == date_str).\
                    all()
                     
                    
dec_temp
```

```python
new_june_temp = []

for temp in june_temp:
    new_june_temp.append(temp.tobs)
    
```

```python
new_dec_temp = []

for temp in dec_temp:
    new_dec_temp.append(temp.tobs)
    
```

```python
# Use t-test to determine if difference in means is statistically significant
# Independent unpaired t test
stats.ttest_ind(new_june_temp, new_dec_temp)
```

```python
# mean is a numpy method
mean(june_temp)
```

```python
# mean is a numpy method
mean(dec_temp)
```

```python
# Discuss which t-test to use and why and what your results mean.
```

#### Unpaired Independent t_test(aka two sample t_test) is performed on mutually exclusive groups to test the differnce is means.
#### It is an unpaired t_test because it compares the means of different group of months(june vs december). 
#### p_value > 0.05 is considered statistically insignificant and difference in means occur by chance.
